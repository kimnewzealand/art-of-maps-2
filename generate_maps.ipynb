```python
import matplotlib.pyplot as plt
import geopandas as gpd
import geodatasets
import contextily as ctx
import matplotlib.pyplot as plt
import osmnx as ox
```

```python
# Create world map with matplotlib and contextily basemap
fig, ax = plt.subplots(figsize=(12,12))
padding=0.5
ax.set_title('World Map')
ax.set_xticks([])
ax.set_yticks([])
ax.set_axis_off()
countries.plot(ax=ax,color='orange',alpha=0.5)
ctx.add_basemap(ax, source=ctx.providers.CartoDB.Positron)
```

```python
# Create interative map with geopandas
countries = gpd.read_file(geodatasets.get_path('naturalearth.land'))
countries.explore()
```

```python
# Create Bike map of Utrecht with matplotlib and osmnx 
place ="Utrecht, Netherlands"
gdf = ox.geocoder.geocode_to_gdf([place])
G = ox.graph.graph_from_place(place, network_type="bike", retain_all=True)
fig, ax = ox.plot.plot_graph(
    G,
    show=False,
    close=False,
    bgcolor="#444444",
    edge_color="#ffa500",
    edge_linewidth=0.3,
    node_size=0,
)
# to this matplotlib axis, add the place shape(s)
gdf.plot(ax=ax, fc="#444444", ec=None, lw=1, alpha=1, zorder=-1)
plt.show()
```
